```python
days=[1,2,3,4,5]

sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='c',label='eating',linewidth=5)
plt.plot([],[],color='r',label='working',linewidth=5)
plt.plot([],[],color='k',label='playing',linewidth=5)

plt.stackplot(days, sleeping,eating,playing,working,colors=['m','c','k',
                                                            'r'])

plt.title('Area plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
slices=[7,7,7,13]
activities=['sleeping','eating','working','playing']
cols=['c','m','r','b']
plt.pie(slices,
       labels=activities,
       colors=cols,
       startangle=90,
       shadow=True,
       explode=(0,0,0.2,0),
       autopct='%1.1f%%')
plt.title('pie chart')
plt.show()
```

```python
import numpy as np
```

```python
def fun(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.subplot(222) 
#plt.subplot(212)
#(222) means we have 2 plot,horizontaly 2 plot and this plot is 2nd plot
#(212) means we have 2 plot,horizontaly 1 plot and this plot is 2nd plot
plt.plot(t1,fun(t1),'bo',t2,fun(t2))

plt.subplot(221)
#plt.subplot(222)
#(221) means we have 2 plot,horizontaly 2 plot and this plot is 1st plot
#(222) means we have 2 plot,horizontaly 2 plot and this plot is 2nd plot
plt.plot(t2,np.cos(2*np.pi*t2))

plt.show()
```
